<span style="color:red;">Here we make the panel data from each merged df from py_controls</span>

```python
import pandas as pd
```

<h1>Panels</h1>

```python
df = pd.read_excel('controls_full_no0/data_reg_performance.xlsx')
df
```

```python
# format y_vars: Ticker Symbol, Date, Metric
y_vars = df.columns[:3].tolist()
x_vars = df.columns[3:].tolist()
```

```python
y_vars
```

```python
# shift for each stock the observations in order to create lags
shifted_x = df.groupby(['Ticker Symbol'])[x_vars].shift(1)

lag_cols = {col: 'Lag ' + col for col in shifted_x.columns}
shifted_x = shifted_x.rename(columns=lag_cols)

shifted_x
```

```python
df = pd.concat([df[y_vars], shifted_x], axis=1)
df
```

```python
# create year dummies and add them
df['Year'] = pd.to_datetime(df['Date']).dt.year

year_dummies = pd.get_dummies(df['Year'], prefix='Year', drop_first=False)
df = pd.concat([df.drop('Year', axis=1), year_dummies], axis=1)
df
```

```python
df.isna().sum()
```

```python
df.to_excel('controls_full_no0/data_reg_performance.xlsx')
```
