```python
import nltk 
import re
from nltk.stem import WordNetLemmatizer
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
from nltk.corpus import words

import pandas as pd
import numpy as np
```

```python
nltk.download("wordnet")
nltk.download('words')
nltk.download('omw-1.4')
```

<h1>Load Lexicon</h1>

```python
def only_az_chars(string):
    return bool(re.match('^[a-z]+$', string))
```

```python
words.fileids()
```

```python
lexicon = words.words('en')
lexicon = pd.DataFrame(lexicon)
lexicon
```

<h1>Lemmatization</h1>

```python
lemmatizer = WordNetLemmatizer()
```

```python
# step 1
lexicon = lexicon[0].str.lower()
```

```python
# step 2
for i in range(0,len(lexicon)):
    string = lemmatizer.lemmatize(lexicon[i])               #lemmatizer.lemmatize("a")
    lexicon[i] = string
```

```python
# step 3
lexicon = np.unique(lexicon)
lexicon
len(lexicon)
```

```python
lexicon = lexicon.tolist()
lexicon
```

```python
allow = [string for string in lexicon if only_az_chars(string)]
allow
```

```python
len(allow)
```
