```python
import pandas as pd
import numpy as np
import re
```

<h1>Making the training set</h1>

```python
def only_az_chars(string):
    return bool(re.match('^[a-z]+$', string))
```

```python
data = pd.read_csv('data_lemma_internet.txt', delim_whitespace=True, header=None)
data
```

```python
data = data.drop(data.columns[[0]], axis=1)
```

```python
data.reset_index()
data.columns = [0,1]
data
```

```python
# convert the labels to absolute count
data[0] = round(data[0] * 181.376)
data
```

```python
arr = data.values
arr = arr.tolist()
arr
```

```python
allow = [set for set in arr if only_az_chars(str(set[1]))]
allow
```

```python
len(allow)
```

```python
pd.DataFrame(allow).to_excel("data_lemma_internet_cleaned.xlsx")
```
